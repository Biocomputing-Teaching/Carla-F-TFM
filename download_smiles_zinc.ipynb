Ens hem baixat les dades de zinc15; 2D, smiles, React.standard. 
Les dades estan en un fitxer on les linies són codis url per baixar els smiles de totes les molècules. El que hem de fer és llegir una per una aquestes linies i descarregar-ho i enganxar-ho en un fitxer. El nostre fitxer de sortida on hi haurà tots els SMILES es diu Smiles_total.csv.

```python
import requests

urls = open("ZINC-downloader-2D-smi.uri","r")
lines = urls.readlines() #llegeix les linies
fh= open("smiles_totalzinc20.csv","a")

for line in lines:
    request = requests.get(line)
    fh.write(request.text)
fh.close()

```
